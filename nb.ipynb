```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_csv('full_execution_times.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

plt.errorbar(df['Hilos'], df['Promedio Tiempo Real (s)'], yerr=df['Error Real (s)'], fmt='-o', label='Tiempo Real', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Usuario (s)'], yerr=df['Error Usuario (s)'], fmt='-o', label='Tiempo Usuario', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Sistema (s)'], yerr=df['Error Sistema (s)'], fmt='-o', label='Tiempo Sistema', capsize=5)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempos de Ejecución vs Número de Hilos')

plt.legend()

plt.show()
```

```python
df = pd.read_csv('step_times.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

plt.errorbar(df['Hilos'], df['Promedio Tiempo Inicialización (s)'], yerr=df['Error Inicialización (s)'], fmt='-o', label='Inicialización', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Asignación (s)'], yerr=df['Error Asignación (s)'], fmt='-o', label='Asignación', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Suma (s)'], yerr=df['Error Suma (s)'], fmt='-o', label='Suma', capsize=5)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempos de Ejecución vs Número de Hilos')

plt.legend()

plt.show()
```
