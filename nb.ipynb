```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('full_execution_times_openmp.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

plt.errorbar(df['Hilos'], df['Promedio Tiempo Real (s)'], yerr=df['Error Real (s)'], fmt='-o', label='Tiempo Real', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Usuario (s)'], yerr=df['Error Usuario (s)'], fmt='-o', label='Tiempo Usuario', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Sistema (s)'], yerr=df['Error Sistema (s)'], fmt='-o', label='Tiempo Sistema', capsize=5)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempos de Ejecución vs Número de Hilos')

plt.legend()

plt.show()
```

```python
df = pd.read_csv('step_times_openmp.csv')

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))

plt.errorbar(df['Hilos'], df['Promedio Tiempo Inicialización (s)'], yerr=df['Error Inicialización (s)'], fmt='-o', label='Inicialización', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Asignación (s)'], yerr=df['Error Asignación (s)'], fmt='-o', label='Asignación', capsize=5)
plt.errorbar(df['Hilos'], df['Promedio Tiempo Suma (s)'], yerr=df['Error Suma (s)'], fmt='-o', label='Suma', capsize=5)

plt.xlabel('Número de Hilos')
plt.ylabel('Tiempo (s)')
plt.title('Tiempos de Ejecución vs Número de Hilos')

plt.legend()

plt.show()
```

```python
hilos_1 = [1.25906, 1.19686, 0.06066]  # Promedios para 1 hilo
error_1 = [0.08723, 0.09011, 0.02408]  # Errores para 1 hilo

hilos_20 = [1.21746, 1.15986, 0.05600]  # Promedios para 20 hilos
error_20 = [0.06745, 0.07085, 0.03193]  # Errores para 20 hilos

labels = ['Tiempo Real', 'Tiempo Usuario', 'Tiempo Sistema']

x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()

bars_1 = ax.bar(x - width/2, hilos_1, width, yerr=error_1, label='1 Hilo con OpenMP', capsize=5)

bars_20 = ax.bar(x + width/2, hilos_20, width, yerr=error_20, label='Sin OpenMP', capsize=5)

ax.set_ylabel('Tiempo (s)')
ax.set_title('Comparación de Tiempos entre 1 y 20 Hilos')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

plt.show()
```
